```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('./data/final.csv')
df.shape
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df = df.drop(columns = ['seats', 'years', 'zip', 'section', 'gender'])
```

```python
sns.pairplot(df)
```

```python
np.abs(df.corr()['renewed']).sort_values(ascending = False)
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(df.season_year, df.renewed)
plt.title('Renewal Rate by Season', size = 20)
plt.xlabel('Season')
plt.ylabel('Renewal Rate')
```

```python
plt.figure(figsize=(10,8))
sns.barplot(df.season_year, df.flash_profit_loss)
plt.title('Resale Profit by Season', size = 20)
plt.xlabel('Season')
plt.ylabel('Resale Profit')
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(df.season_year, df.attendance)
plt.title('Attendance by Season', size=20)
plt.xlabel('Season')
plt.ylabel('Attendance')
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(df.season_year, df.percent_total_used)
plt.title('Tickets Used by Season', size=20)
plt.xlabel('Season')
plt.ylabel('Tickets Used')
```

```python
np.abs(df.corr()[['renewed']])
# np.fill_diagonal(corrs.values, -2)
df.corr()[['renewed']]
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(np.abs(df.corr()[['renewed']]).sort_values(by = 'renewed', ascending = False), annot = True, cmap = 'coolwarm')
plt.title('Correlations to Renewal', size=20)
```
