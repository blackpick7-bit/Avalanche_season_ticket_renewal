```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GRU
from tensorflow.keras.optimizers import Adam


import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/final.csv')
df.drop(columns = ['gender', 'years', 'zip', 'seats', 'section'], inplace = True)
df.columns
```

```python
train = df[df['season_year'] !='2018-2019'].drop(columns = ['season_year'])
test = df[df['season_year'] == '2018-2019'].drop(columns = ['season_year'])
```

```python
X_train = train.drop(columns = ['renewed'])
X_test = test.drop(columns = ['renewed'])
y_train = train['renewed']
y_test = test['renewed']
```

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
n_input = X_train.shape[1]
```

```python
model = Sequential()
model.add(Dense(21, input_dim = n_input, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['acc']
)
```

```python
# Fit it
history = model.fit(X_train, y_train,
         validation_data=(X_test, y_test),
         epochs = 100, 
         batch_size=32,
         verbose=2)
```

```python
train_loss = history.history['loss']
test_loss = history.history['val_loss']

plt.figure(figsize=(12,8))
plt.plot(train_loss, label = 'Training Loss', color = 'navy')
plt.plot(test_loss, label = 'Test Loss', color = 'skyblue')
plt.legend()
```

```python
model_dropout = Sequential()
model_dropout.add(Dense(17, input_dim = n_input, activation = 'relu'))
model_dropout.add(Dropout(rate = .5))
model_dropout.add(Dense(15, activation = 'relu'))
model_dropout.add(Dropout(rate = .5))
model_dropout.add(Dense(1, activation = 'sigmoid'))

model_dropout.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["acc"]
)

history_dropout = model_dropout.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=32,
    verbose=2
)
```
