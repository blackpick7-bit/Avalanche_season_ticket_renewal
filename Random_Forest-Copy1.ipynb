```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier 
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
df = pd.read_csv('./data/features.csv')
df.columns
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df.columns
```

```python
df.renewed.value_counts(normalize=True)
```

```python
train = df[df['season_year'] !='2018-2019'].drop(columns = ['season_year'])
test = df[df['season_year'] == '2018-2019'].drop(columns = ['season_year'])
```

```python
train
```

```python
test.shape
```

```python
X_train = train.drop(columns = ['renewed'])
X_test = test.drop(columns = ['renewed'])
y_train = train['renewed']
y_test = test['renewed']
```

```python
y_train.value_counts(normalize = True)
```

```python
rf = RandomForestClassifier(random_state=42)
rf_params = {'n_estimators':[12, 13, 14, 15], 'max_depth':[5], 'max_features':[None, 15, 14]}
gs = GridSearchCV(rf, rf_params, cv = 5)
gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)
```

```python
gs.score(X_test, y_test)
```

```python
preds = gs.predict(X_test)
```

```python
proba = gs.predict_proba(X_test)
```

```python
listo = []
for each in proba:
    listo.append(each[1])
np.mean(listo)
```

```python
accuracy_score(y_test, preds)
```

```python
lr = LogisticRegressionCV()
lr.fit(X_train, y_train)
lr.score(X_train, y_train)
# lr.score(X_test, y_test)
```

```python
lr.score(X_test, y_test)
```

```python
preds = list(lr.predict(X_test))
```

```python
preds
```

## random train test split

```python
X = df.drop(columns = ['renewed', 'season_year'])
y = df['renewed']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, stratify = y)
```

```python
lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
gs.fit(X_train,y_train)
print(gs.best_score_)
print(gs.best_params_)
```

```python
preds = gs.predict(X_test)
```

```python
accuracy_score(y_test, preds)
```

```python
vote = VotingClassifier([
    ('ada', AdaBoostClassifier(base_estimator=DecisionTreeClassifier())),
    ('gb', GradientBoostingClassifier()),
    ('tree', DecisionTreeClassifier()),
    ('forest', RandomForestClassifier())
])

vote_params = {
    'ada__n_estimators': [30, 40],
    'ada__base_estimator__max_depth': [10],
    'forest__n_estimators':[13, 14, 15],
    'forest__max_depth':[7],
    'forest__max_features':[None, 16]
    
}

gs = GridSearchCV(vote, param_grid=vote_params, cv=3)
gs.fit(X_train, y_train)
print(gs.best_score_)
gs.best_params_
```

```python
gs.score(X_test, y_test)
```
