### Max seat

```python

def max_seat(season):
    listo = []
    for each in season['seats']:
        if each == []: 
            listo.append(0)
        else:   
            listo.append(max(each))
    season['seat_end'] = listo

max_seat(df2015)
max_seat(df2016)
max_seat(df2017)
max_seat(df2018)
max_seat(df2019)
```

### Min seat

```python

def min_seat(season):
    listo = []
    for each in season['seats']:
        if each == []: 
            listo.append(0)
        else:   
            listo.append(min(each))
    season['seat_start'] = listo

min_seat(df2015)
min_seat(df2016)
min_seat(df2017)
min_seat(df2018)
min_seat(df2019)
```

## Creating distance

```python
def find_zip(contact, data, season):
    zip2 = []
    for each in contact:
        if len(str(data[data['contact_id'] == each]['zip'].unique()[0])) ==5:
            zip2.append(data[data['contact_id'] == each]['zip'].unique()[0])
        elif len(str(data[data['contact_id'] == each]['zip'].unique()[0])) < 5:
            zip2.append(0)
        else:
            zip2.append(str(data[data['contact_id'] == each]['zip'].unique()[0])[:5])

    season['zip'] = zip2  
    season['zip'] = season['zip'].astype(str)
```

```python
find_zip(contacts2015, data2015, df2015)
find_zip(contacts2016, data2016, df2016)
find_zip(contacts2017, data2017, df2017)
find_zip(contacts2018, data2018, df2018)
find_zip(contacts2019, data2019, df2019)
```

```python
df2016.zip
```

```python
#code from Ed Prentice
from uszipcode import SearchEngine as ZipcodeSearchEngine
from geopy.distance import vincenty

def distance(data):
    data['zip'] = data['zip'].str[:5]
    data['zip'] = data['zip'].fillna('')
    data['distance'] = 0.0

    search = ZipcodeSearchEngine()

    for i in range(len(data)):
        if data['zip'][i] != '':

            zipcode = search.by_zipcode(data['zip'][i])
            long = zipcode.lng
            lat = zipcode.lat
            zipped = (lat, long)
            pep_long_lat = (39.7487, -105.0077)
            data['distance'][i] = vincenty(pep_long_lat, zipped).miles

        else:
            data['distance'][i] = np.nan
```

```python
distance(df2015)
distance(df2016)
distance(df2017)
distance(df2018)
distance(df2019)
```

```python
def sections(data):    
    num_sect = []
    multiple = []
    for x in data['sections']:
        if len(x) > 1:
            num_sect.append(len(x))
            multiple.append(1)
        else:
            num_sect.append(1)
            multiple.append(0)
    data['num_sections'] = num_sect
    data['multiple_sections'] = multiple

sections(df2015)
sections(df2016)
sections(df2017)
sections(df2018)
sections(df2019)
```
